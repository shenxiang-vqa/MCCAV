```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

att_map = np.load('att_map_deep_global.npz')
att = att_map['img_sa'].squeeze(0)

# 8
fig, axs = plt.subplots(8, 1, figsize=(5, 20))

for i in range(8):
    # i
    b = att[i:i+1, :, :].squeeze(0)

    # 
    sns.heatmap(b, cmap='PuRd', vmin=0, vmax=1, cbar=True, square=True, xticklabels=True, yticklabels=True, ax=axs[i])

    # 
    axs[i].set_title(f'Attention Head {i+1}')

    # 
    x_interval = 4
    y_interval = 4
    axs[i].set_xticks(np.arange(0, b.shape[1], step=x_interval))
    axs[i].set_xticklabels(np.arange(0, b.shape[1], step=x_interval))
    axs[i].set_yticks(np.arange(0, b.shape[0], step=y_interval))
    axs[i].set_yticklabels(np.arange(0, b.shape[0], step=y_interval))

# 
plt.tight_layout()

# 
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

att_map = np.load('att_map_deep_global.npz')
att = att_map['img_ga'].squeeze(0)
print(att.shape)

# 8
fig, axs = plt.subplots(8, 1, figsize=(18, 20))

for i in range(8):
    # i
    b = att[i:i+1, :, :].squeeze(0)

    # 
    sns.heatmap(b, cmap='PuRd', vmin=0, vmax=1, cbar=True, square=True, xticklabels=True, yticklabels=True, ax=axs[i], cbar_kws={'aspect': 20})

    # 
    axs[i].set_title(f'Attention Head {i+1}')

    # 
    x_interval = 4
    y_interval = 4
    axs[i].set_xticks(np.arange(0, b.shape[1], step=x_interval))
    axs[i].set_xticklabels(np.arange(0, b.shape[1], step=x_interval))
    axs[i].set_yticks(np.arange(0, b.shape[0], step=y_interval))
    axs[i].set_yticklabels(np.arange(0, b.shape[0], step=y_interval))

# 
plt.tight_layout()

# 
plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

att_map = np.load('att_map_deep_global.npz')
att = att_map['img_ga'].squeeze(0)
c = att[1:2, :, :].squeeze(0)
print(c.shape)

# 
fig, ax = plt.subplots(figsize=(5, 5))

# 
im = ax.imshow(c, cmap='PuRd', vmin=0, vmax=1, aspect='auto')

# 
y_interval = 4
ax.set_yticks(np.arange(0, c.shape[0], step=y_interval))
ax.set_yticklabels(np.arange(0, c.shape[0], step=y_interval))

# 
ax.set_xticks(np.arange(0, c.shape[1]))
ax.set_xticklabels(np.arange(0, c.shape[1]))

# 
ax.set_frame_on(False)

# 
ax.tick_params(axis='x', colors='m')
ax.tick_params(axis='y', colors='r')

# 
save_dir = 'result'
os.makedirs(save_dir, exist_ok=True)

# result
plt.savefig(os.path.join(save_dir, 'attention_head_3_4.png'), dpi=600)

# 
#plt.subplots_adjust(bottom=0.4, right=0.35, top=0.6)

# 
plt.show()
```
